### Q1 During this period, seven companies offered scooters. How many scooters did each company have in this time frame? Did the number for each company change over time? Did scooter usage vary by company?

````python
# Total number of sooters each company has
scooternum = '''
SELECT COUNT(Distinct sumdid), companyname
FROM scooters
WHERE sumdgroup <> 'bicycle'
GROUP BY companyname;
'''


with engine.connect() as connection:
    scount = pd.read_sql(text(scooternum), con = connection)

scount
````

````python
# Number of sooters each company has in May
snum_may = '''
SELECT COUNT(distinct sumdid), companyname
FROM scooters
WHERE sumdgroup <> 'bicycle'
AND pubdatetime < '2019-06-01'
GROUP BY companyname;
'''


with engine.connect() as connection:
    scount_may = pd.read_sql(text(snum_may), con = connection)

scount_may
````

````python
# Number of sooters each company has in June
snum_june = '''
SELECT COUNT(distinct sumdid), companyname
FROM scooters
WHERE sumdgroup <> 'bicycle'
AND pubdatetime < '2019-07-01'
GROUP BY companyname;
'''


with engine.connect() as connection:
    scount_june = pd.read_sql(text(snum_june), con = connection)

scount_june
````

````python
# Number of sooters each company has in July
snum_july = '''
SELECT COUNT(distinct sumdid), companyname
FROM scooters
WHERE sumdgroup <> 'bicycle'
AND pubdatetime < '2019-08-01'
GROUP BY companyname;
'''


with engine.connect() as connection:
    scount_july = pd.read_sql(text(snum_july), con = connection)

scount_july
````
